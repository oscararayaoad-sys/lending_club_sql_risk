## LendingClub Dataset — Documentación de Columnas

### Préstamo (Loan)
| Columna | Descripción |
|---------|-------------|
| `loan_amnt` | Monto solicitado por el prestatario (principal requested) |
| `funded_amnt` | Monto realmente financiado (puede ser < que loan_amnt) |
| `term` | Plazo del préstamo: 36 o 60 meses |
| `issue_d` | Fecha de emisión del préstamo |

### Riesgo (Risk)
| Columna | Descripción |
|---------|-------------|
| `int_rate` | Tasa de interés anual (0.1299 = 12.99%) |
| `grade` | Calificación general de riesgo: A=mejor, G=peor |
| `sub_grade` | Calificación más fina: A1-A5, B1-B5, ..., G5 |
| `loan_status` | **Estado del préstamo: Fully Paid, Charged Off (defaulteó), Current, Late** ⭐ TARGET |

### Prestatario (Borrower)
| Columna | Descripción |
|---------|-------------|
| `annual_inc` | Ingreso anual auto-reportado en USD |
| `emp_length` | Años de antigüedad laboral: 0=<1 año, 10=10+ años |
| `home_ownership` | Tipo de vivienda: RENT, OWN, MORTGAGE, OTHER |
| `addr_state` | Estado/provincia de residencia (códigos 2 letras) |

### Crédito & Deuda (Credit)
| Columna | Descripción |
|---------|-------------|
| `dti` | Debt-to-income ratio: (pagos mensuales deuda) / (ingreso mensual). LC acepta si < 35% |
| `delinq_2yrs` | Número de veces 30+ días en mora en últimos 2 años. 0=nunca |

*Se hipotiza que 'emp_length' figura con tal cantidad de nulos por el dato incluyente de signos como ">, <, +, etc". Lo que para python es un string y no un object. Se procede a comprobar.*

```python
print(df_a_limpiar['emp_length'].dropna().nunique()) #Cantidad de las variables sin nulos.
print('---')
print(df_a_limpiar['emp_length'].value_counts(dropna=False)) #Cualidad de las variables con nulos (NaN).
```

*Se confirma presencia de strings en una columna tipo floar. Tambien se observan nulos reales dentro de la columna. Se procede a remapear columna 'emp_length' y limpar nulos.*

```python
df_a_limpiar = df_a_limpiar.dropna(subset=['emp_length']) #Descarte de nulos.

emp_length_map = { #Remapeo.
    '< 1 year': 0,
    '1 year': 1,
    '2 years': 2,
    '3 years': 3,
    '4 years': 4,
    '5 years': 5,
    '6 years': 6,
    '7 years': 7,
    '8 years': 8,
    '9 years': 9,
    '10+ years': 10,
}
df_a_limpiar['emp_length'] = df_a_limpiar['emp_length'].map(emp_length_map)
```

```python
#Comprobación.
print(df_a_limpiar['emp_length'].dropna().nunique()) #Cantidad de las variables.
print('---')
print(df_a_limpiar['emp_length'].value_counts()) #Cualidad de las variables.
```

*Se decarta las entradas con 'loan_status' sin resultado definitivo (Pagado o incobrable). Dado que el objetivo es saber que predice que un prestamo caiga en mora, se deben descartar aquellos casos que no aporten información concreta. Tambien se reasignan variables para prestamos pagados y no pagados (good y bad) y se empieza con una nueva tabla.*

```python
good = ['Fully Paid', 'Does not meet the credit policy. Status:Fully Paid'] #Pagados.
bad = ['Charged Off', 'Does not meet the credit policy. Status:Charged Off'] #Incobrables.

df_modelo = df_a_limpiar[df_a_limpiar['loan_status'].isin(good + bad)].copy() #Excluimos todo lo que no necesitamos.
df_modelo['default'] = df_modelo['loan_status'].isin(bad).astype(int) #Asignamos valor a la nueva columna 'default'.

print(df_modelo['default'].value_counts())
print('Filas: ',df_modelo.shape[0])
print(f'Tasa de default:  {df_modelo['default'].mean():.1%}')
```

*Se chequean valores atipicos en columnas 'dti' y 'annual_inc' porque se entiende que 'annual_inc' es un dato que el mismo cliente dio, y 'dti' es un resultado en donde 'annual_inc' fue utilizado en la ecuación.*

```python
print('-----annual_inc-----')
print(df_modelo['annual_inc'].describe().apply(lambda x: f'{x:,.2f}'))
print('-----dti-----')
print(df_modelo['dti'].describe().apply(lambda x: f'{x:,.2f}'))
```

*Se investigan valores atipicos.*

```python
print('annual_inc > 1M: ', (df_modelo['annual_inc'] > 1_000_000).sum()) #excluimos ingresos anuales mayor a 1M. Si bien pueden ser reales, no son objetivo de prestamos. 
print('annual_inc < 1K: ', (df_modelo['annual_inc'] < 1_000).sum()) #excluimos ingresos anuales menores a 1K. Ingresos menores se atribuyen a error de formato o de entrada.
print('dti negativo: ', (df_modelo['dti'] < 0).sum()) #dti es un porcentaje de LendingClub. No puede ser menor a 0.
print('dti > 100: ', (df_modelo['dti'] > 100).sum()) #dti es un porcentaje de LendingClub. No puede ser mayor a 100.
```

*Se sospecha correlación.*

```python
ingreso_alto = df_modelo['annual_inc'] > 1_000_000
dti_alto = df_modelo['dti'] > 100

print('Filas con ambos problemas: ', (ingreso_alto & dti_alto).sum())
print('Filas con ingreso alto: ', (ingreso_alto & -dti_alto).sum())
print('Filas con dti alto: ', (dti_alto &  -ingreso_alto).sum())
```

*Se descarta correlación comprobando solapamiento. Se procede a descartar outliers.*

```python
#Quitamos outliers.
df_modelo = df_modelo[
    (df_modelo['annual_inc'] <= 1_000_000) &
    (df_modelo['annual_inc'] >= 1000) &
    (df_modelo['dti'] >= 0) &
    (df_modelo['dti'] <= 100)
].copy()

print('-----Filas post-limpieza de outliers-----')
print(df_modelo.shape[0])
print('-----annual_inc-----')
print(df_modelo['annual_inc'].describe().apply(lambda x: f'{x:,.2f}'))
print('-----dti-----')
print(df_modelo['dti'].describe().apply(lambda x: f'{x:,.2f}'))
```
